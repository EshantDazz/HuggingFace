```python
# instruct models

LLAMA = "meta-llama/Meta-Llama-3.1-8B-Instruct"
PHI3 = "microsoft/Phi-3-mini-4k-instruct"
GEMMA2 = "google/gemma-2-2b-it"
QWEN2 = "Qwen/Qwen2-7B-Instruct" # exercise for you
MIXTRAL = "mistralai/Mixtral-8x7B-Instruct-v0.1" # If this doesn't fit it your GPU memory, try others from the hub
```

```python
messages = [
    {"role": "system", "content": "You are a helpful assistant"},
    {"role": "user", "content": "Tell a light-hearted joke for a room of Data Scientists"}
  ]
```

```python
# Quantization Config - this allows us to load the model into memory and use less memory

quant_config = BitsAndBytesConfig(
    load_in_4bit=True,  #Load 4 bits instead of 32 bits
    bnb_4bit_use_double_quant=True,  #Quantize twice not once
    bnb_4bit_compute_dtype=torch.bfloat16,  #in calculations use this bfloat16 makes some improvement
    bnb_4bit_quant_type="nf4" #representation of 4 bit numbers
)
```

```python
# Tokenizer

tokenizer = AutoTokenizer.from_pretrained(LLAMA)
tokenizer.pad_token = tokenizer.eos_token #which token is to fill up the prompt if there is something to be added later then we use this
inputs = tokenizer.apply_chat_template(messages, return_tensors="pt").to("cuda")
```

```python
# The model

model = AutoModelForCausalLM.from_pretrained(LLAMA, device_map="auto", quantization_config=quant_config)
```

THis model is the actual LLM model and is actually the series of pytorch layers

```python
memory = model.get_memory_footprint() / 1e6
print(f"Memory footprint: {memory:,.1f} MB")
```

```python
model
```

You can see the layers such as embedding, self attention layer and towards the end they use Silu Activation Function

```python
outputs = model.generate(inputs, max_new_tokens=80)
```

```python
print(tokenizer.decode(outputs[0]))
```

8 billion Llama model and we double quantized it to 4 bits

```python
# Clean up

del inputs, outputs, model
torch.cuda.empty_cache()
```

```python
# Wrapping everything in a function - and adding Streaming

def generate(model, messages):
  tokenizer = AutoTokenizer.from_pretrained(model)
  tokenizer.pad_token = tokenizer.eos_token
  inputs = tokenizer.apply_chat_template(messages, return_tensors="pt").to("cuda")
  streamer = TextStreamer(tokenizer)
  model = AutoModelForCausalLM.from_pretrained(model, device_map="auto", quantization_config=quant_config)
  outputs = model.generate(inputs, max_new_tokens=80, streamer=streamer)
  del tokenizer, streamer, model, inputs, outputs
  torch.cuda.empty_cache()
```

```python
generate(PHI3, messages)
```

```python
generate(QWEN2, messages)
```
